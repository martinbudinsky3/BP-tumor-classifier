```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
data = pd.read_csv('../datasets/centromeres.txt', sep='\t')

data
```

```python
centromeres_data = data[data['type'] == 'centromere']

centromeres_data
```

```python
data_without_cols = centromeres_data.loc[:, 'chrom':'chromEnd']

data_without_cols
```

```python
data_with_col_names = data_without_cols.copy()

data_with_col_names.rename(columns = { 'chrom': 'Chromosome', 'chromStart': 'Start', 'chromEnd': 'End' }, inplace = True)

data_with_col_names
```

```python
data_without_chr = data_with_col_names.copy()

data_without_chr.iloc[:, 0] = data_without_chr.loc[:, 'Chromosome'].apply(lambda x: x.replace('chr', ''))

data_without_chr
```

```python
final_centromeres_data = data_without_chr.reset_index(drop=True)

final_centromeres_data
```
